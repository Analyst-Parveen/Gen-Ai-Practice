```python
from langchain_huggingface import HuggingFaceEmbeddings

embedding=HuggingFaceEmbeddings(
    model_name='sentence-transformers/all-MiniLM-L6-v2')

text='delhi is the capital of india'

vector=embedding.embed_query(text)

# print(vector)
print(len(vector))
```

```python
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
```

```python
doc1 = Document(
    page_content=(
        "Virat Kohli is one of the best batsmen in modern cricket. "
     
    ),
    metadata={
        "ipl_team": "Royal Challengers Bangalore",
        "role": "Batsman"
    }
)

doc2 = Document(
    page_content=(
        "Rohit Sharma is an elegant batsman and a successful captain in limited-overs cricket. "
        
    ),
    metadata={
        "ipl_team": "Mumbai Indians",
        "role": "Batsman"
    }
)

doc3 = Document(
    page_content=(
        "Jasprit Bumrah is one of the best fast bowlers in the world. "
        
    ),
    metadata={
        "ipl_team": "Mumbai Indians",
        "role": "Bowler"
    }
)

doc4 = Document(
    page_content=(
        "MS Dhoni is a legendary cricketer known for his calm leadership and finishing ability. "
     
    ),
    metadata={
        "ipl_team": "Chennai Super Kings",
        "role": "Wicketkeeper-Batsman"
    }
)

doc5 = Document(
    page_content=(
        "Hardik Pandya is a dynamic all-rounder who contributes with both bat and ball. "
    
    ),
    metadata={
        "ipl_team": "Gujarat Titans",
        "role": "All-Rounder"
    }
)
```

```python
doc=[doc1,doc2,doc3,doc4,doc5]
```

```python
vector_store=Chroma(
    embedding_function=embedding,
    persist_directory='MY_CHROMA_DB',
    collection_name='sample'
)
   
```

```python
# adding data in vector_store
vector_store.add_documents(doc)
```

```python
# view vector_store
vector_store.get(include=['embeddings','documents','metadatas'])
```

```python
# vestor_search
vector_store.similarity_search(query='who among these are a batsman',k=3)
```

```python
# vestor_search_with_score
vector_store.similarity_search_with_score(query='who among these are a batsman',k=1)
```

```python
# meta-data-filtering
vector_store.similarity_search_with_score(query='',filter={'ipl_team':'Chennai Super Kings'},k=2)
```

```python
# update_document
doc6 = Document(
    page_content=(
        "Ravindra Jadeja is a world-class all-rounder known for his sharp left-arm spin bowling. "
       
    ),
    metadata={
        "ipl_team": "Chennai Super Kings",
        "role": "All-Rounder"
    }
)
vector_store.update_document(document=doc6,document_id='c74eeeaa-83a0-4425-a385-da892be98c16')
```

```python
vector_store.get(include=['embeddings','documents','metadatas'])
```
