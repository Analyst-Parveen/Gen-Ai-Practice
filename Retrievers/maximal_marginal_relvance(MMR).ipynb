```python
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embedding=HuggingFaceEmbeddings(
    model_name='sentence-transformers/all-MiniLM-L6-v2')
```

```python
doc=[Document(page_content='His name is Parveen sharma'),
    Document(page_content='people call him parveen'),
    Document(page_content='His name in document his also parveen sharma'),
    Document(page_content='His learns AI'),
    Document(page_content='His hobby is playing volleyball'),
    Document(page_content='He is male'),
    Document(page_content='His gender was also male in his documents'),
    Document(page_content='He was Msc Passout'),
    Document(page_content='He wake up early in the morning'),
    Document(page_content='His want people call him parveen.') 
     ]
```

```python
vectorstore=Chroma.from_documents(
    documents=doc,
    embedding=embedding,
    persist_directory='MMR'
)
```

```python
Retrievers=vectorstore.as_retriever(search_type='mmr',search_kwargs={'k':5,'lambda_mult':0.5})
```

```python
query='tell me about the person'
```

```python
result=Retrievers.invoke(query)
```

```python
for i,doc in enumerate(result):
    print(f'Result ---{i+1}')
    print(doc.page_content)
```
