```python
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embedding=HuggingFaceEmbeddings(
    model_name='sentence-transformers/all-MiniLM-L6-v2')


```

```python
doc1 = Document(
    page_content=(
        "Virat Kohli is one of the best batsmen in modern cricket. "
     
    ),
   
)

doc2 = Document(
    page_content=(
        "Rohit Sharma is an elegant batsman and a successful captain in limited-overs cricket. "
        
    ),

)

doc3 = Document(
    page_content=(
        "Jasprit Bumrah is one of the best fast bowlers in the world. "
        
    ),

)

doc4 = Document(
    page_content=(
        "MS Dhoni is a legendary cricketer known for his calm leadership and finishing ability. "
     
    ),

)

doc5 = Document(
    page_content=(
        "Hardik Pandya is a dynamic all-rounder who contributes with both bat and ball. "
    
    ),

)
```

```python
doc=[doc1,doc2,doc3,doc4,doc5]
```

```python
vector_store=Chroma.from_documents(documents=doc,
                                   embedding=embedding,
                                   
                                   persist_directory='collection')

```

```python
retrievers=vector_store.as_retriever(search_kwargs={'k':2})
```

```python
query="who is bats man among all"
```

```python
docs=retrievers.invoke(query)
```

```python
for i,doc in enumerate(docs):
    print(f'result ---={i+1}')
    print(doc)
```
